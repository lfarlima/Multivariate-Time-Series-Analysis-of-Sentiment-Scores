```python
import os
import pandas as pd
from newsapi import NewsApiClient
from dotenv import load_dotenv
load_dotenv()
```

```python
api_key = os.getenv("news_api")
```

```python
newsapi = NewsApiClient(api_key=api_key)
```

```python
# pltr_headlines = newsapi.get_top_headlines(q="Palantir", language="en", country="us")
# pltr_headlines


pltr_headlines = newsapi.get_everything(
    q="Palantir",
    language="en",
    page_size=100,
    sort_by="relevancy",
    from_param="2021-02-28"
)

pltr_headlines
```

```python
 # Transform the response dictionary to a DataFrame
pltr_df = pd.DataFrame.from_dict(pltr_headlines["articles"])
```

```python
pltr_df.tail()
```

```python
pltr_df.to_csv("pltr_march29.csv", index=True, encoding='utf-8-sig')
```
